```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
##Imagens para carregar
img1 = cv.imread('Imagens/regua/casa315f.jpg',0)
img2 = cv.imread('Imagens/regua/casa315g.jpg',0)
#cv.__version__
```

```python
##Iniciando o detector SIFT
#sift = cv.xfeatures2d.SIFT_create()
#sift = cv.xfeatures2d.SIFT_create()
#http://answers.opencv.org/question/199318/how-to-use-sift-in-python/
#orb = cv.ORB_create(edgeThreshold=5)
orb = cv.ORB_create(nfeatures=2000, scoreType=cv.ORB_FAST_SCORE)
#orb = cv.ORB_create(scoreType=cv.ORB_FAST_SCORE)
##Encontrando os pontos chaves com SIFT
#kp1 = orb.detect(img1,None)
#kp2 = orb.detect(img2,None)

kp = orb.detect(img2,None)

# marcador BF com parametros padrão
kp, des = orb.compute(img2, kp)
img3 = cv.drawKeypoints(img2,kp,None,color=(0,255,0), flags=0)
cv.imwrite('Imagens/regua/casa315gORB.jpg',img3)
plt.imshow(img3),plt.show()
#img3 = draw_matches(img1,kp1,img2,kp2,good)
```
