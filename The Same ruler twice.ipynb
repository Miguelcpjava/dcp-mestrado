```python
#Importando as bibliotecas necess√°rias
import numpy as np
import cv2 
from matplotlib import pyplot as plt
```

```python
img1 = cv2.imread('Imagens/regua/casa315g.jpg',cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('Imagens/regua/casa315gCopia.jpg',cv2.IMREAD_GRAYSCALE)
```

```python
#Inicializando o ORB
orb = cv2.ORB_create(nfeatures=9000, scoreType=cv2.ORB_FAST_SCORE,edgeThreshold=25, patchSize=25)
#surf = cv2.xfeatures2d.SURF_create(400)
```

```python
# find the keypoints and descriptors with SIFT
kp1, des1 = orb.detectAndCompute(img1,None)
kp2, des2 = orb.detectAndCompute(img2,None)
```

```python
# create BFMatcher object
#bf = cv2.BFMatcher(cv2.NORM_L1,crossCheck=True)
#bf = cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=False)
bf = cv2.BFMatcher()
```

```python
# Match descriptors.
matches = bf.match(des1,des2)
```

```python
# Sort them in the order of their distance.
matches = sorted(matches, key = lambda x:x.distance)

# Draw first 10 matches.
img5 = cv2.drawMatches(img1,kp1,img2,kp2,matches[:50],None, flags=2)

plt.imshow(img5),plt.show()
cv2.imwrite('Imagens/regua/casa315gMATCHESTwo.jpg',img5)
```
