```python
#Importando as bibliotecas necess√°rias
import numpy as np
import cv2 
from matplotlib import pyplot as plt
```

```python
# Pegando as imagens para o matcher
img1 = cv2.imread('Imagens/regua/casa315g.jpg',0)
img2 = cv2.imread('Imagens/regua/casa315g1.jpg',0)
img3 = cv2.imread('Imagens/regua/casa315g2.jpg',0)
img4 = cv2.imread('Imagens/regua/casa315g3.jpg',0)
```

```python
#Inicializando o ORB
#orb = cv2.ORB_create(nfeatures=1500, scoreType=cv2.ORB_FAST_SCORE)
orb = cv2.ORB_create(nfeatures=1500, scoreType=cv2.ORB_FAST_SCORE,edgeThreshold=20, patchSize=20)
```

```python
# find the keypoints and descriptors with SIFT
kp1, des1 = orb.detectAndCompute(img1,None)
kp2, des2 = orb.detectAndCompute(img2,None)
kp3, des3 = orb.detectAndCompute(img3,None)
kp4, des4 = orb.detectAndCompute(img4,None)
```

```python
# create BFMatcher object
#bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
#https://docs.opencv.org/3.0-beta/doc/py_tutorials/py_feature2d/py_matcher/py_matcher.html
bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=False)
```

```python
# Match descriptors.
matches = bf.match(des1,des4)
```

```python
# Sort them in the order of their distance.
matches = sorted(matches, key = lambda x:x.distance)

# Draw first 10 matches.
img5 = cv2.drawMatches(img1,kp1,img4,kp4,matches[:50],None, flags=2)

plt.imshow(img5),plt.show()
cv2.imwrite('Imagens/regua/casa315gMATCHES3.jpg',img5)
```
