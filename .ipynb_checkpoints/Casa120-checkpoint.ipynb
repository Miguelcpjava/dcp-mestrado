```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img1 = cv2.imread('Imagens/regua/Casa315g.jpg',cv2.IMREAD_GRAYSCALE)
#img2 = cv2.imread('Imagens/regua/Casa120a.jpg',cv2.IMREAD_GRAYSCALE)
#img2 = cv2.imread('Imagens/regua/Casa120b.jpeg',cv2.IMREAD_GRAYSCALE)
img2 = cv2.imread('Imagens/regua/Casa120.jpeg',cv2.IMREAD_GRAYSCALE)
```

```python
#orb = cv2.ORB_create(nfeatures=1500, scoreType=cv2.ORB_FAST_SCORE,edgeThreshold=25, patchSize=25)
orb = cv2.ORB_create(nfeatures=1500,scoreType=cv2.ORB_FAST_SCORE,edgeThreshold=15, patchSize=31,fastThreshold=20)
```

```python
kp1, des1 = orb.detectAndCompute(img1,None)
kp2, des2 = orb.detectAndCompute(img2,None)
```

```python
bf = cv2.BFMatcher(cv2.NORM_HAMMING,crossCheck=False)
```

```python
matches = bf.match(des1,des2)
# Sort them in the order of their distance.
matches = sorted(matches, key = lambda x:x.distance)

good = []
for m,n in matches:
    if m.distance < 0.7*n.distance:
        good
# Draw first 10 matches.
img5 = cv2.drawMatches(img1,kp1,img2,kp2,matches[:50],None, flags=2)

plt.imshow(img5),plt.show()
cv2.imwrite('Imagens/regua/casa120aTwo.jpg',img5)
```
